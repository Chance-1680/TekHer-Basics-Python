### Exercise 26:

### Exercise 27:
Create a radar chart with 5 variables and random values between 0 and 1.


```python
# Exercise 27
labels = ['A', 'B', 'C', 'D', 'E']
values = np.random.rand(5)
angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)
values = np.concatenate((values, [values[0]]))
angles = np.concatenate((angles, [angles[0]]))
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, values, 'o-', linewidth=2)
ax.fill(angles, values, alpha=0.25)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
plt.title("Radar Chart")
plt.show()
```

### Exercise 44:
Create a waterfall chart using Matplotlib to show the changes in a sequence of values.


```python
# Exercise 44
values = [100, -30, 50, -10, 60]
changes = np.cumsum([0] + values)
fig, ax = plt.subplots()
ax.bar(range(len(values)), values, bottom=changes[:-1])
plt.title("Waterfall Chart")
plt.show()
```
